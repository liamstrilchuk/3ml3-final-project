```python
from model import Model

model = Model()
model.create(drop_data=0, dropout=0.3, max_features=5000, layer_size_factor=0.25)
model.train(epochs=20)
```

```python
def print_sorted_report(model):
	report_dict = model.get_report(output_dict=True, threshold=0.7)
	report_dict = sorted(report_dict.items(), key=lambda i: i[1]["f1-score"], reverse=True)

	for proj, scores in report_dict:
		print(f"{proj:<30} {round(scores['precision'], 3):<5} {round(scores['recall'], 3):<5} {round(scores['f1-score'], 3):<5}")

print_sorted_report(model)
```

```python
model.optimize_thresholds(step=0.01, beta=0.5)
print_sorted_report(model)
```
