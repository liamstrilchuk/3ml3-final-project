```python
# count the number of occurrences of a word in the dataset
def count_words(word):
	total, with_word = 0, 0
	with open("../data/all_data.csv", "r") as f:
		import csv
		reader = csv.reader(f)
		for line in reader:
			total += 1
			if word in line[1].lower():
				with_word += 1
	return total, with_word

total, with_the = count_words("the")
total, with_canadian = count_words("canadian")

round(with_the / total * 100, 2), round(with_canadian / total * 100, 2)
```
